```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('House Price India.csv')
df.info()
```

```python
# Univariate analysis
sns.countplot(data=df,x='number of floors')
```

```python
sns.scatterplot(data = df,x = df.index,y = "Renovation Year",hue = "Built Year")
```

```python
plt.boxplot(data = df,x = "Number of schools nearby")
```

```python
plt.hist(df['Price'])
```

```python
# Bivariate analysis

sns.scatterplot(data=df,x='number of bedrooms',y='Price')
```

```python
sns.boxplot(data = df,x ="Area of the house(excluding basement)",y = 'Price')
```

```python
sns.lineplot(x = df[df['number of bedrooms'] == 2]['condition of the house'], y = df[df['number of bedrooms'] == 2]['Price'])
```

```python
sns.violinplot(data= df,x="number of bedrooms",y="Price")
```

```python
sns.barplot(data= df,x = "condition of the house",y ="Price")
```

```python
# Multivariate

sns.heatmap(df.corr())
```
